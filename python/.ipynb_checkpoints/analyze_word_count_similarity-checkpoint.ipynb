```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

speech_type = 'so'

so_results = np.load(f"{speech_type}_similarities.npy")

x = np.array(so_results[10])
y = np.arange(x.shape[0])
color = "blue" if speech_type == 'so' else 'red'
plt.xlabel("sentence")
plt.ylabel("non-similarity")
plt.plot(y, x, color=color)
figure = plt.gcf()
figure.set_size_inches(20, 10)
plt.show()
```

```python
signal = np.sin(2*np.pi*x/x.shape[0])
F = np.fft.fft(signal)
F_abs = np.abs(F)
figure = plt.gcf()
figure.set_size_inches(20, 10)
plt.plot(F_abs)
```
