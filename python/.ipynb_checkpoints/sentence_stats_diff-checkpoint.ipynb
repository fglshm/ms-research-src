```python
import pandas as pd
from statistics import mean, median, variance, stdev

def calculate_stats(lst, speech_type):
    standing_ovation = 1 if speech_type == 'so' else 0
    mean_value = round(mean(lst), 2)
    median_value = round(median(lst), 2)
    variance_value = round(variance(lst), 2)
    stdev_value = round(stdev(lst), 2)
    return [standing_ovation, mean_value, median_value, variance_value, stdev_value]

save_filename = 'all_sentence_stats.csv'
save_df = pd.DataFrame(
        columns=[
            "title",
            "standing ovation",
            "mean",
            "median",
            "variance",
            "standard deviation"
        ]
    )
speech_types = ['nso', 'so']
idx = 0
for speech_type in speech_types:
    read_filename = "{}_sentence_stats.csv".format(speech_type)
    df = pd.read_csv(read_filename)
    columns = df.columns
    all_id = df['id']
    all_mean = df['mean']
    all_median = df['median']
    all_variance = df['variance']
    all_stdev = df['standard deviation']
    all_sentence_count = df['sentence count']
    
    _mean = calculate_stats(all_mean, speech_type)
    _mean.insert(0, 'mean')
    
    _median = calculate_stats(all_median, speech_type)
    _median.insert(0, 'median')
    
    _variance = calculate_stats(all_variance, speech_type)
    _variance.insert(0, 'variance')
    
    _stdev = calculate_stats(all_stdev, speech_type)
    _stdev.insert(0, 'standard deviation')
    
    _sentence_count = calculate_stats(all_sentence_count, speech_type)
    _sentence_count.insert(0, 'sentence count')
    
    save_df.loc[idx] = _mean
    idx += 1
    save_df.loc[idx] = _median
    idx += 1
    save_df.loc[idx] = _variance
    idx += 1
    save_df.loc[idx] = _stdev
    idx += 1
    save_df.loc[idx] = _sentence_count
    idx += 1
#     print("--- {} ---".format(speech_type))
#     print(calculate_stats(all_mean))
#     print(calculate_stats(all_median))
#     print(calculate_stats(all_variance))
#     print(calculate_stats(all_stdev))
#     print(calculate_stats(all_sentence_count))
    
save_df.to_csv(save_filename, index=False)
```
